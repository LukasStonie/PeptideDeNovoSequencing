```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
pools = ['Pool_49', 'Pool_52', 'Pool_60']
```

```python
plot = plt.figure()
for pool in pools:
        # Load the data
        raw_data =pd.read_csv(
                f"../../Data/Datasets/{pool}/Thermo_SRM_{pool}_01_01_3xHCD-1h-R2-tryptic/msmsScans.txt",
                sep='\t', index_col=None)
        plt.hist(raw_data['m/z'], bins=100, alpha=0.3, label=' '.join(pool.split('_')))
plt.xlabel('m/z')
plt.ylabel('Frequency')
plt.title('Histogram of m/z values')
plt.legend()
plt.show()
```
